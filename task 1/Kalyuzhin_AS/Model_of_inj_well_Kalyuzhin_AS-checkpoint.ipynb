```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.integrate import odeint
import json
```

# Receiving data block

```python
#----------json block----------------

input_data = '{"gamma_w": 1, "h":2030, "d_in":129, "q_liq":364, "eps":7.4E-5, "p_wh": 19.9, "MD": [0, 1015, 2030], "TVD": [0, 950.5, 1643], "T0": 26.3}'
exc = json.loads(input_data)

#-------------------------------------

#---------data block------------------

gamma_w = exc["gamma_w"] # specific gravity of water
H = exc["h"] # deepness of well in m
d_m = exc["d_in"] * 1e-03
q_m3day = exc["q_liq"] # inner diameter in m
eps = exc["eps"] # roughness
p_wh = exc["p_wh"] * 0.1 # wellhead pressure in MPa

MD = np.array(exc["MD"])
TVD = np.array(exc["TVD"])

T_0 = exc["T0"] # temperature in C
TempGrad = 0.02463 # temperature gradient deg C per 100 m (and it's valid for Kelvins because of dT_C = dT_K)

#----------------------------------------
```

# Functions block

```python
# функция расчета плотности воды в зависимости от давления и температуры
def rho_w_kgm3(P_Mpa,T_K):
    rho_w_sc_kgm3 = 1000
    return rho_w_sc_kgm3 /(1+(T_K-273)/10000*(0.269*(T_K-273)**(0.637)-0.8))
    
# Расчет вязкости воды в зависимости от температуры и давления
def visc_w_cP(P_Mpa,T_K):
    A = 109.574
    B = 1.1217
    muw = A * (1.8 * T_K - 460) ** (-B) * (0.9994 + 0.0058 * P_Mpa + 0.6534 * (10) ** (0 - 4) * P_Mpa * P_Mpa)
    return muw

# Расчет числа Рейнольдса 
def Re(q_m3day, d_m, mu_mPas = 0.2, rho_kgm3 = 1000):
    # q_m3day - дебит жидкости, м3/сут
    # rho_kgm3 - плотность воды или жидкости, по умолчанию 1000 кг/м3, чистая вода
    # mu_mPas  - вязкость жидкости по умолчанию 0.2 мПас
    # d_m      - диаметр трубы, м
    v_ms = q_m3day/ 86400 / 3.1415 * 4 / d_m ** 2
    return rho_kgm3 * v_ms * d_m / mu_mPas * 1000

def friction(q_m3day, eps, d_m1 = d_m, mu_mPas = 0.2, rho_kgm3 = gamma_w * 1000):
    Re_val = Re(q_m3day, d_m1, mu_mPas, rho_kgm3)
    if Re_val < 3000:
        return 64/Re_val
    else:
        return 1/(1.14-2 * np.log10(eps/d_m + 21.25 / (Re_val ** 0.9)))**2

# функция расчета градиента давления в нагнетательной скважин
def pressure_gradient_MPam(q_m3day, eps, P_Mpa, T_C, d_m1 = d_m, cos_alpha =1):
    # q_m3day - дебит жидкости, м3/сут
    # P_Mpa - давление, МПа
    # T_C - температура, С
    # d_m - диаметр 
    # Ws - соленость 
    # cos_alpha - косинус угол отклонения от вертикали 
    # roughness_m - шероховатость
    rho_kgm3 = rho_w_kgm3(P_Mpa, T_C + 273)
    mu_cP = visc_w_cP(P_Mpa, T_C + 273)
    f = friction(q_m3day, eps, d_m1, mu_cP, rho_kgm3)
    g = 9.81
    q_m3sec = q_m3day /86400
    return (rho_kgm3 * g * cos_alpha - 0.815 * f * rho_kgm3 /( d_m ** 5) * (q_m3sec )**2) / 1000000
```

# Data preprocessing block:

```python
#------interpolation of data-------------

# cubic interpolation can't be used because we should have at least 4 calculating points
# linear interpolation used instead

Udl_m = interp1d(MD, MD-TVD, kind='linear') # elongation from measured deep - for plotting graphics
Hv_m = interp1d(MD,TVD, kind='linear') # vertical deep from measured

cos_ang = np.diff(TVD)/np.diff(MD) # array of cosines
cos_ang_comp = np.insert(cos_ang, 0, 1.0) # inserting of 1 on first position 
cos_interp = interp1d(MD, cos_ang_comp, kind = 'linear', fill_value='extrapolate') # interpolation of cosines array


#----------------------------------------------

#---------------plotting block-----------------

h_ = np.linspace(0, H, num=100, endpoint=True)

plt.plot(Udl_m(h_), Hv_m(h_), '-')
plt.xlabel("Deviation from the vertical, m")
plt.ylabel("Well deepness , m")
plt.title("Well trajectory scheme")
ax = plt.gca()
ax.invert_yaxis()
plt.show()

plt.plot(cos_interp(h_), (h_), '-')
plt.xlabel("Cosine of deviation angle")
plt.ylabel("Well deepness , m")
plt.title("Cosine of deviation angle")
ax = plt.gca()
ax.invert_yaxis()
plt.show()
#-----------------------------------------------
```

Solution of equations for pressure and temperature distribution in wellbore. It may be solved as system of two equations of the form:

$$ \frac{dP}{dL} = \frac{1}{10^{5}} \left [  \rho g  \cos \alpha  - 0.815 \frac{f \rho}{d^5} q ^ 2  \right ]  $$

$$ \frac{dT}{dL} = geograd $$

Boundary conditions set at the wellhead

# Solution and results block:

```python
def dPTdL(PT, h):
    # rounding to 5 decimal places needs because it's necessary to avoid accumulation of machine rounding errors
    dPdL = round(pressure_gradient_MPam(q_m3day, eps, PT[0], PT[1], d_m, cos_interp(h)), 5) 
    dTdL = round(TempGrad * cos_interp(h), 5)
    return [dPdL, dTdL]

PTwh = [p_wh,T_0] # boundary conditions

hs = np.linspace(0, H, 200) # area of integration and step

PTs = odeint(dPTdL, PTwh, hs) # numerically solving of system of differential equations

# saving of results
P = PTs[:,0]
T = PTs[:,1]

#--------------plotting results block-------------------

plt.plot(P,hs, label ="pressure")
plt.plot(T,hs, label = "temperature")
plt.xlabel("P, Т")
plt.ylabel("h, measured well deepness, m")
ax = plt.gca()
ax.invert_yaxis()
plt.legend()
plt.title("Pressure distribution");
plt.show()

plt.plot(P,Hv_m(hs), label ="pressure")
plt.plot(T,Hv_m(hs), label = "temperature")
plt.xlabel("P, Т")
plt.ylabel("h, measured well deepness, m")
ax = plt.gca()
ax.invert_yaxis()
plt.legend()
plt.title("Pressure distribution");
plt.show()

#----------------------------------------------
```

# Calculation of dependence of inlet pressure from debit

```python
Q_var = np.linspace(10, 500, 10)
def dPTdL_1(PT, h, q):
    dPdL = round(pressure_gradient_MPam(q, eps, PT[0], PT[1], d_m, cos_interp(h)), 5)
    dTdL = round(TempGrad * cos_interp(h), 5)
    return [dPdL, dTdL]

Ps = ([])
for q_var in Q_var:
    PTs = odeint(dPTdL_1, PTwh, hs, (q_var,))
    P = np.amax(PTs[:,0], 0)
    Ps = np.append(Ps, P)
```

Plotting block:

```python
plt.plot(Q_var, Ps)
plt.xlabel("Q, m3/day")
plt.ylabel("P, MPa")
plt.title("Dependence of inlet pressure from debit");
plt.show()
```

# Calculation of dependence of inlet pressure from outlet temperature

```python
def dPTdL_2(PT, h):
    dPdL = round(pressure_gradient_MPam(q_m3day, eps, PT[0], PT[1], d_m, cos_interp(h)), 5)
    dTdL = round(TempGrad * cos_interp(h), 5)
    return [dPdL, dTdL]

Ps_2 = ([])
T_var = np.linspace(10, 50, 10)
for T in T_var:
    PTwh_var = [p_wh, T]
    PTs_2 = odeint(dPTdL_2, PTwh_var, hs)
    P_2 = np.amax(PTs_2[:,0], 0)
    Ps_2 = np.append(Ps_2, P_2)
```

Plotting block:

```python
plt.plot(T_var, Ps_2)
plt.xlabel("T, deg. Celsius")
plt.ylabel("P, MPa")
plt.title("Dependence of inlet pressure from outlet temperature");
plt.show()
```

# Creating the .json file block

```python
P_mod = list(PTs[:,0])

q_l = list(Q_var)
p_q = list(Ps)
t = list(T_var)
p_t = list(Ps_2)

frame = {"ex1" : {"h" : [hs[0], hs[(len(hs) - 1)//5], hs[2 * (len(hs) - 1)//5], hs[3 * (len(hs) - 1)//5], hs[4 * (len(hs) - 1)//5], hs[len(hs) - 1]], 
                 "p" : [P_mod[0], P_mod[(len(P_mod) - 1)//5], P_mod[2 * (len(P_mod) - 1)//5], P_mod[3 * (len(P_mod) - 1)//5], P_mod[4 * (len(P_mod) - 1)//5], P_mod[len(P_mod) - 1]],
                },
         "ex2" : {"q_liq" : q_l, "p" : p_q},
         "ex3" : {"t" : t, "p" : p_t}              
        }

json.dump(frame , open(r'C:\Users\Компьютер\Desktop\Kalyuzhin_AS\data.json','w+'))
```
